### Taylor Approximation

The idea behind Taylor Approximation is simple:
we want to approximate any function by a polynomial. First, we need to choose the place where we want
to approximate. In this lab, we will use $x = 0$ for simplicity. And then, we want given function and our approximation
to have the same derivatives at the chosen point. (You will explore Taylor series in much more detail in
Calculus 2)

Let $f(x)=\frac{2}{x+1}$

At first, we want our approximation to have the same value at $x = 0$ as $f(x)$, which is 2 (since $f(0)=2$ ) . So let $T0(x) = 2$. This is our first approximation. Let's plot $f(x)$ and $T0(x)$

```python
f=2/(x+1)
fn=sp.lambdify(x,f)
T0=2
```

```python
fig=plt.figure()

xval=np.linspace(-0.99, 2, 1000)

plt.plot(xval, fn(xval),color='blue', label='f(x)')
plt.axhline(T0, color='red', label='T0(x)')  # note it is easier to plot constant functions using horizontal line command

plt.axvline(0,color='black', lw=2)
plt.axhline(0,color='black', lw=2)

plt.ylim(-1,10)


plt.grid()
plt.legend()
plt.title("Degree-0 approximation")
plt.show()
```

Next, we will add a linear term to our approximation in such a way that first derivatives of $f(x)$ and of
our approximation are the same and the values of our function and its approximation are also the same.
Here are the steps:

i: Find derivative of $f(x)$ at $x = 0$.

ii: We want $T1(x) = T0(x) + ax$ for some $a$. Note that $T1(0)=f(0)$, so the values are the same at $x=0$. However, we donâ€™t know what $a$ is. So, find derivative of $T1(x)$ and make it
equal to derivative of $f(x)$ at $x = 0$. This will give you the value for $a$.

iii: Now define $T1(x) = T0(x) + ax$ replacing $a$ with the value you found in last part.

iv: Finally, plot $f(x)$ and $T1(x)$ on the same graph. You should notice that this is just a linear approximation (tangent line).


```python
f.diff(x).subs(x,0)
```

```python
T1=T0+a*x
T1.diff(x)
```

So this means $a=-2$.

```python
T1=T0+(-2)*x
T1
```

```python
fig=plt.figure()
t1n=sp.lambdify(x,T1)
xval=np.linspace(-0.99, 2, 1000)


plt.plot(xval, fn(xval),color='blue', label='f(x)')
plt.plot(xval, t1n(xval),color='red', label='T1(x)')

plt.axvline(0,color='black', lw=2)
plt.axhline(0,color='black', lw=2)

plt.ylim(-2,10)


plt.grid()
plt.legend()
plt.title("Degree-1 approximation")
plt.show()
```

Next, we will add a quadratic term and make sure second derivatives are the same as well as first derivatives and the values. So, as in the previous step:

i: Find second derivative of $f(x)$ at $x = 0$.

ii: We want $T2(x) = T1(x) + bx^2$, but we donâ€™t know what $b$ is. So, find the second derivative of $T2(x)$ and make it equal to the second derivative of $f(x)$ at $x = 0$. This will give you the value for $b$. (Also note, that $T2(0)=f(0)$ and $T2'(0)=f'(0)$).

iii: Now define $T2(x) = T1(x) + bx^2$
replacing $b$ with the value you found in last part.

iv: Finally, plot $f(x)$ and $T2(x)$ on the same graph.

```python
f.diff(x,2).subs(x,0)
```

```python
T2=T1+b*x**2
T2.diff(x,2)
```

So here we have that $2b=4$, so $b=2$.

```python
T2=T1+(2)*x**2
T2
```

```python
fig=plt.figure()
t2n=sp.lambdify(x,T2)
xval=np.linspace(-0.99, 2, 1000)

plt.plot(xval, fn(xval),color='blue', label='f(x)')
plt.plot(xval, t2n(xval),color='red', label='T2(x)')

plt.axvline(0,color='black', lw=2)
plt.axhline(0,color='black', lw=2)

plt.ylim(-1,10)

plt.grid()
plt.legend()
plt.title("Degree-2 approximation")
plt.show()
```

Next, we find cubic approximation $T3(x)=T2(x)+cx^3$

```python
f.diff(x,3).subs(x,0)
```

```python
T3=T2+c*x**3
T3.diff(x,3)
```

So, $6c=-12$ and therefore $c=-2$

```python
T3=T2+(-2)*x**3
T3
```

```python
fig=plt.figure()
t3n=sp.lambdify(x,T3)
xval=np.linspace(-0.99, 2, 1000)

plt.plot(xval, fn(xval),color='blue', label='f(x)')
plt.plot(xval, t3n(xval),color='red', label='T3(x)')

plt.axvline(0,color='black', lw=2)
plt.axhline(0,color='black', lw=2)

plt.ylim(-1,10)

plt.grid()
plt.legend()
plt.title("Degree-3 approximation")
plt.show()
```

And one more time for quartic approximation:

```python
f.diff(x,4).subs(x,0)
```

```python
T4=T3+d*x**4
T4.diff(x,4)
```

So we can see that $d=2$.

```python
T4=T3+(2)*x**4
T4
```

```python
fig=plt.figure()
t4n=sp.lambdify(x,T4)
xval=np.linspace(-0.99, 2, 1000)

plt.plot(xval, fn(xval),color='blue', label='f(x)')
plt.plot(xval, t4n(xval),color='red', label='T4(x)')

plt.axvline(0,color='black', lw=2)
plt.axhline(0,color='black', lw=2)

plt.ylim(-1,10)

plt.grid()
plt.legend()
plt.title("Degree-4 approximation")
plt.show()
```

We can repeat these steps as high as we need to.

Note that the way we did here works only for approximation at $x=0$. If you want to try making them at other places, we need a small modification, which we will show below. 

*Note: this is not needed for the lab. It is only if you want to play around with this*.

So, suppose we want to approximate $f(x)$ and $x=p$, then:

$T0(x)=f(p)$

$T1(x)=T0(x)+a(x-p)$

$T2(x)=T1(x)+b(x-p)^2$

$T3(x)=T2(x)+c(x-p)^3$

and so on.

### Linear Approximation and Differentials

One of the problems in the Lab will be about differentials and linear approximation. If you didn't cover this in class, here are main things you need to know before starting the problem.

Suppose you are given a function $f(x)$ and a point $x$. Let $\Delta x$ be a small change in the $x$. Then we define $\Delta y = f(x+\Delta x)-f(x)$. [Note: we often would call $\Delta x$ as change in $x$, and $\Delta y$ as change in $y$]

Next we define differentials. The differential $dx = \Delta x$, and differential $dy=f'(x)dx$. 

The problem itself will guide you and explain the meaning of $dy$, but as a quick explanation: $dy$ is an error of linear approximation (i.e Taylor Approximation with linear function). In other words, how much the linear approximation differs from the actual value.
